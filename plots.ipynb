```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def get_points(file_path):
    xs=[]
    ys=[]
    with open(file_path) as f:
        for line in f:
            x, y = line.split()
            xs.append(x)
            ys.append(y)
    xs=np.array(xs).astype(float)
    ys=np.array(ys).astype(float)
    return xs,ys
    
```

# Varying the value of w

```python

x_rk_0_4,y_rk_0_4 = get_points('cte_0.4/rk_0_4.txt')
x_rk_0_8,y_rk_0_8 = get_points('cte_0.8/rk_0_8.txt')
x_rk_1_2,y_rk_1_2 = get_points('cte_1.2/rk_1_2.txt')
x_rk_1_6,y_rk_1_6 = get_points('cte_1.6/rk_1_6.txt')
x_rk_2,y_rk_2 = get_points('cte_2/rk_2.txt')


```

```python
plt.plot(x_rk_0_4,y_rk_0_4,color='blue',label='0.4')
plt.plot(x_rk_0_8,y_rk_0_8,color='red',label='0.8')
plt.plot(x_rk_1_2,y_rk_1_2,color='green',label='1.2')
plt.plot(x_rk_1_6,y_rk_1_6,color='orange',label='1.6')
plt.legend()
plt.show()
```
